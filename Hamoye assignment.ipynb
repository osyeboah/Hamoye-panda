```python
# Question 8

# Perform a groupby operation on 'Nationality' and calculate the sum for each area
sum_by_nationality = df.groupby('Nationality')['Overall'].sum()

# Sort the sums in ascending order
sorted_sums = sum_by_nationality.sort_values()

# Find the 7th lowest sum
seventh_lowest_sum = sorted_sums.iloc[6]  # Indexing starts from 0, so 6 corresponds to the 7th element

# Find the nationality with the 7th lowest sum
nationality_with_seventh_lowest_sum = sorted_sums.index[sorted_sums.values == seventh_lowest_sum][0]

print("Nationality with the 7th lowest sum in 2017:", nationality_with_seventh_lowest_sum)
print("The 7th lowest sum in 2017:", seventh_lowest_sum)
```
